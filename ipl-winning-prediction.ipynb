```python
pd.set_option('display.max_columns',500)

match = pd.read_csv('/kaggle/input/ipl-data-set/matches.csv')
delivery = pd.read_csv('/kaggle/input/ipl-data-set/deliveries.csv')
```

```python
match.head()
```

```python
delivery.head()
```

```python
total_score = delivery.groupby(["match_id","inning"])['total_runs'].sum().reset_index()
```

```python
total_score_1_inning = total_score[total_score['inning'] == 1]
```

```python
match_df = match.merge(total_score_1_inning[["match_id","total_runs"]],left_on='id',right_on='match_id')
```

```python
match_df.head()
```

```python
match_df.team1.unique()
```

```python
 teams = [
            'Sunrisers Hyderabad',
            'Mumbai Indians',
            'Royal Challengers Bangalore',
            'Kolkata Knight Riders',
            'Kings XI Punjab',
            'Chennai Super Kings',
            'Rajasthan Royals',
            'Delhi Capitals'
        ]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied'] == 0]

match_df = match_df[['match_id','city','winner','total_runs','venue']]
```

```python
delivery_df = match_df.merge(delivery,on = 'match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.head()
```

```python
delivery_df['current_score']= delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df['run_left'] = delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df['ball_left'] = 126-delivery_df['over']*6-delivery_df['ball']
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x== "0" else "1")
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)
delivery_df['wicket_left']=10- delivery_df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120-delivery_df['ball_left'])
```

```python
delivery_df['rrr'] = (delivery_df['run_left']*6)/delivery_df['ball_left']
```

```python
delivery_df['result'] = np.where((delivery_df['winner'] == delivery_df['batting_team']),1,0 )
```

```python
delivery_df
```

```python
delivery_df = delivery_df[['batting_team','bowling_team','city','run_left','ball_left','wicket_left','total_runs_x','crr','rrr','result']]
```

```python
delivery_df = delivery_df.dropna()
```

```python
delivery_df = delivery_df.sample(delivery_df.shape[0])
```

```python
delivery_df
```

```python
delivery_df = delivery_df[(delivery_df['ball_left'] != 0)]
delivery_df = delivery_df[(delivery_df['run_left'] > 0)]
```

```python
final_df = delivery_df.copy()
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer(
    [('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])]
                      ,remainder='passthrough' )
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps = [
        ('step1',trf),
            ('step2',LogisticRegression(solver='liblinear'))]
               )
```

```python
pipe.fit(X_train,y_train
        )
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
